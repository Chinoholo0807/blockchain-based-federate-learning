```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
def draw_plot(x,y,xlabel='round',ylabel='acc',learning_rate=0.01,
                n_client=10,n_participators=5,epochs=5,batch_size=10,
                model_name='mnist_2nn',aggreagate_method='fed_avg',attacker_prop=0.0):
    plt.plot(x,y)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    title = 'federate learning:n_client({}) n_participators({}) epochs({}) lr({})batch_size({}) model_name({}) aggregate({}) attaker({}%)'.format(
        n_client,
        n_participators,
        epochs,
        learning_rate,
        batch_size,
        model_name,
        aggreagate_method,
        attacker_prop
    )
    plt.title(title)
    plt.show()
```

```python
input_str = 'aaaddddfffdddaddd'

def check(start,end,str):
    while start <= end:
        if str[start] != str[end]:
            return False
        start = start +1
        end = end -1
    return True

n = len(input_str)
maxlen = 0
for i in range(0,n):
    for j in range(i,n):
        result = check(i,j,input_str)
        if result:
            maxlen = max(maxlen,j-i+1)
print(maxlen)
```
